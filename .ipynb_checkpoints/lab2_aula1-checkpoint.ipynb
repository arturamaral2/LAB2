```python
import pandas as pd 
import numpy as np
```

```python
def analise_txt(nome):
    df = pd.read_csv(nome, header = None)
    df = df.rename(columns = {0:'tempo'})
    df['tempo_anterior'] = df['tempo'].shift(1)
    df['diferenca_tempo'] = df['tempo'] - df['tempo_anterior']
    media = df['diferenca_tempo'].mean()
    desvio_padrao = df['diferenca_tempo'].std()
    df['diferenca_tempo'].hist()
    print(f'{nome} \n')
    print('Media :', media )
    print('Desvio :', desvio_padrao)
    
    
    
```

```python
analise_txt('Tempo_sem_100ms.txt')
```

```python
analise_txt('Tempo_sem_10ms.txt')
```

```python
analise_txt('Tempo_sem_1ms.txt')
```

## tempo real 

```python
analise_txt('Tempo_com_100ms.txt')
```

```python
analise_txt('Tempo_com_10ms.txt')
```

```python
analise_txt('Tempo_com_1ms.txt')
```
